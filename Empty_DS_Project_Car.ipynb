```python
## Fetch cars whose price is >12000

c['price'] = pd.to_numeric(c['price'], errors='coerce')

cars_above_12000 = c.query('price > 12000')
cars_above_12000
```

Note:
error, it might be due to the data type of the 'price' column.
If the 'price' column contains strings instead of numeric values, the comparison might fail.

When errors='coerce' is used:

1. It instructs pandas to convert problematic or non-convertible entries into NaN (Not a Number) values.
2. If pandas encounters any data that it can't convert to a numeric type,
instead of raising an error, it converts that specific value to NaN and continues
the conversion process for the rest of the data.
