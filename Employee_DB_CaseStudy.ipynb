```python
yearly_total_pay = sal.groupby('Year')['TotalPay'].sum()

# Calculate percentage distribution
percentage_distribution = (yearly_total_pay / yearly_total_pay.sum()) * 100

# Plotting the percentage distribution
plt.figure(figsize=(8, 6))
percentage_distribution.plot(kind='bar')
plt.title('Percentage Distribution of Yearwise TotalPay')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
